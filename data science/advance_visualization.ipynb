````python
import pandas as pd
````

````python
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
````

````python
df = sns.load_dataset('penguins') # df is a dataframe (excel like sheet)
````

````python
df
````

````python
fig = px.line(df, x='culmen_length_mm',)
````

````python
fig.write_html('penguins_length_line.html',auto_open=True)
````

````python
fig = px.scatter(df,x='culmen_length_mm',y='culmen_depth_mm',color='island')
fig.write_html('penguins_scatter.html',auto_open=True)
````

````python
df.dropna(inplace=True)
````

````python
fig = px.scatter(df,x='culmen_length_mm',
                 y='culmen_depth_mm',
                 color='species',
                 size='body_mass_g')
fig.write_html('penguins_scatter.html',auto_open=True)
````

````python
fig = px.scatter_3d(df,x='culmen_length_mm',
                 y='culmen_depth_mm',
                 z='body_mass_g',
                 symbol='sex',
                 color='species',
                 size='body_mass_g')
fig.write_html('penguins_scatter.html',auto_open=True)
````

````python
fig = px.area(df,x='species',y='body_mass_g',color='sex')
fig.write_html('penguins_sunburst.html',auto_open=True)
````

````python
planets = sns.load_dataset('planets')
planets
````

````python
fig = px.area(planets,x='mass',y='distance')
fig.write_html('planets_area.html',auto_open=True)
````

````python
sns.set(rc={'figure.figsize':(6,5)})
ax= sns.countplot(x='species',data=df)
for x,bar in enumerate(ax.patches):
    y = bar.get_height()
    plt.text(x,2,str(y),horizontalalignment='center',fontdict={'color':'white'})
````

````python
ax = sns.distplot(planets.distance)
````

````python
ax = sns.distplot(df.culmen_length_mm)
````

````python
ax = sns.jointplot(x='mass',y='distance',data=planets,)
````

````python
ax = sns.pairplot(df)
````

````python
ax = sns.pairplot(df,hue='species')
````
