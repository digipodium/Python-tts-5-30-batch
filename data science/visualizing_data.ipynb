```python
x = np.arange(-6,+6, .2)
y1 = np.sin(x)
plt.plot(x,y1,'ro--')
plt.title("the sine wave")
plt.xlabel('T values')
plt.ylabel('sine')
plt.show()
```

```python
x = np.arange(-6,+6, .2)
x2 = x ** 2
x3 = x ** 3
plt.plot(x,'-',label='original')
plt.plot(x2,'--',label='doubled')
plt.plot(x3,':',label='tripled')
plt.legend()
plt.show()
```

```python
# marks 
names= ['ram','shyam','balram','ghanshyam']
marks = [90,54,23,79]
```

```python
plt.plot(marks,'ro:')
plt.xticks(ticks=[0,1,2,3],labels=names,rotation=90)
plt.show()
```

```python
plt.bar(x=names,height=marks)
plt.show()
```

```python
plt.bar(x=names,height=marks)
plt.title('comparison of marks ')
plt.xlabel('brothers')
plt.ylabel('marks')
plt.text(1,90,'Ram is better performer',fontdict={'size':10,'color':'blue'})
plt.show()
```

```python
plt.bar(x=names,height=marks,color=['orange','red'])
plt.title('comparison of marks ')
plt.xlabel('brothers')
plt.ylabel('marks')
plt.text(1,90,'Ram is better performer',fontdict={'size':10,'color':'blue'})
plt.show()
```

```python
plt.bar(x=names,
        height=marks,
        hatch='\\',
        width=.3,
        fill=False,
        edgecolor=['red','green'])
plt.title('comparison of marks ')
plt.xlabel('brothers')
plt.ylabel('marks')
plt.text(1,90,'Ram is better performer',fontdict={'size':10,'color':'blue'})
plt.show()
```

```python
rainfall = [200,290,144,110,189]
cities= ['lucknow','kanpur','banaras','jhansi','allahabad'] 
plt.barh(cities,rainfall)
# lets look at pie
```

```python
rainfall = [200,290,144,110,189]
cities= ['lucknow','kanpur','banaras','jhansi','allahabad'] 
plt.pie(rainfall,
        labels=cities,
        autopct='%.2f%%')
plt.show()
```

```python
rainfall = [200,290,144,110,189]
cities= ['lucknow','kanpur','banaras','jhansi','allahabad'] 
plt.pie(rainfall,
        labels=cities,
        autopct='%.2f%%',
        explode = [0,.1,0,0,0])
plt.show()
```

```python
rainfall = [200,290,144,110,189]
cities= ['lucknow','kanpur','banaras','jhansi','allahabad'] 
plt.pie(rainfall,
        labels=cities,
        autopct='%.2f%%',
        explode = [0,.1,0,0,0],
        shadow=True,
        colors=['red','green','blue','yellow','orange'])
plt.show()
```

```python
rainfall = [200,290,144,110,189]
cities= ['lucknow','kanpur','banaras','jhansi','allahabad'] 
plt.pie(rainfall,
        labels=cities,
        autopct='%.2f%%',
        explode = [0,.1,0,0,0],
        shadow=True,
        wedgeprops={
            'width':.4
        },
        pctdistance=.8,
        radius=1.35)
plt.show()
```
